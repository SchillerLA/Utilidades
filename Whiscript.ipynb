```python
### PREMISSAS
Arquivo_ou_URL = "https://www.instagram.com/reel/DVJfY-0DdaN/?igsh=MWluZmN1NG5neW42bw==" #@param {type:"string"}
Idioma = "pt" #@param ["pt", "en", "es"]
Legenda_automatica_se_disponível = True # @param {type:"boolean"}

### CÓDIGO
def install_whiscript():
    try:
        import whiscript
    except ImportError:
        print("Primeiro uso -> instalando bibliotecas...")
        !pip install "git+https://bdc12:{TOKEN}@github.com/bdc12/whiscript.git" --quiet
import whiscript
from whiscript import transcrever, TranscriberConfig
from google.colab import files
install_whiscript()

cfg = TranscriberConfig(idioma=Idioma, preferir_legenda_automatica=Legenda_automatica_se_disponível)
try:
  txt = transcrever(Arquivo_ou_URL, config=cfg)
  files.download(txt)
except:
  print("Erro na transcrição.")
```
